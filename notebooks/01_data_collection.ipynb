## Create cleaned starter datasets

```python
rent_apr_2026 = rents_clean[rents_clean["Time period"] == "2026-04-01"].copy()

rent_apr_2026 = rent_apr_2026.rename(columns={
    "Time period": "date",
    "Area code": "area_code",
    "Area name": "area_name",
    "Rental price": "average_monthly_rent",
    "Region or country name": "region"

})

rent_apr_2026.head()
```

```python
house_prices_apr_2026 = house_prices[house_prices["Date"] == "2026-04-01"].copy()

house_prices_apr_2026 = house_prices_apr_2026.rename(columns={
    "Date": "date",
    "Region_Name": "area_name",
    "Area_Code": "area_code",
    "Region or country name": "region",
    "Average_Price": "average_house_price"

})

house_prices_apr_2026 = house_prices_apr_2026[
    ["date", "area_code", "area_name", "average_house_price"]
]

house_prices_apr_2026.head()
```

```python
income_clean = income.rename(columns={
    "local authority: district / unitary (as of April 2023)": "area_name",
    "Annual pay - gross": "median_annual_pay",
    "Conf %": "income_confidence_pct"
})

income_clean["median_annual_pay"] = pd.to_numeric(
    income_clean["median_annual_pay"],
    errors="coerce"
)

income_clean.head()
```

```python
rent_apr_2026.to_csv(PROCESSED_DATA_DIR / "rent_apr_2026.csv", index=False)
house_prices_apr_2026.to_csv(PROCESSED_DATA_DIR / "house_prices_apr_2026.csv", index=False)
income_clean.to_csv(PROCESSED_DATA_DIR / "income_2025.csv", index=False)
```
